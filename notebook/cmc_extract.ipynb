```python
import sys
import os
# Add root path

# Thêm root folder (project) vào sys.path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))
from config.logger_config import LoggerConfig
from config.variable_config import COINMARKETCAP_CONFIG
```

```python
headers = {
    "Accept": "application/json",
    "X-CMC_PRO_API_KEY": COINMARKETCAP_CONFIG["cmc_api"]
}
```

```python
import requests

response = requests.get(COINMARKETCAP_CONFIG["cmc_url"], headers=headers)
data = response.json()
```

```python
import json
with open("/home/duc/coin_flow_big_data/data/raw/raw.json","w") as f:
    json.dump(data,f,indent=4)

```

```python
full_data = []
for d in data["data"]:
    data_test = {
        "name":d["name"],
        "symbol":d["symbol"],
        "price":d["quote"]["USD"]["price"],
        "volume_24h":d["quote"]["USD"]["volume_24h"],
        "percent_change_1h":d["quote"]["USD"]["percent_change_1h"],
        "percent_change_24h":d["quote"]["USD"]["percent_change_24h"],
        "percent_change_7d":d["quote"]["USD"]["percent_change_7d"],
        "last_updated":d["quote"]["USD"]["last_updated"],
        "market_cap_dominance":d["quote"]["USD"]["market_cap_dominance"],
        "circulating_supply":d["circulating_supply"]
    }
    full_data.append(data_test)
```

```python
import pandas as pd

df = pd.DataFrame(full_data,index=None)
df
```
